# I. Introduction

#### What is the context of the work?

Economic inequality is a huge problem in the United States. Wealth distribution is uneven and there are wealth gaps between racial groups. For this project I took the bottom 20% average household income of White, White Alone Not Hispanic, Asian, Hispanic, and Black households and compared them to the top 5% average household income of the same racial groups' households over the years 2002 to 2019. I want to see how they compare to each other annd to the GDP per capita over the years, which is theoretically a measure of a persons income.

#### What research questions are you trying to answer?
 - Is the GDP per capita usually close to a specific racial group's average income?
 - Is there a bigger disparity between the top and bottom households in racial groups or between racial groups overall?
 - Is economic inequality getting better?

#### What are your main findings?

 - Economic inequality is getting worse because the rich are getting richer and the poor are stagnant.
 - GDP per capita is decreasing and is closer to the bottom 20% household incomes than the top 5%.
 - We can assume that the top 5% hoseholds do consistently increase every year, while we can't estimate whether bottom 20% households will increase or decrease.
 - There is more inequality between racial groups than inside racial groups.
 
Overall, the implications is that racial wealth inequality is a real and major problem in the United States.

Next, I want to see how each group compares to the GDP per capita numerically by computing the ratios and seeing how they compare to each other. 

```python
#function to compute ratios
def get_gdp_ratio(column, gdp_column):
    gdp_ratios = []
    for value in gdp_column: 
        for i in column:
            ratio=value/i
            if ratio>=1:
                ratio = int(ratio)
            else:
                ratio=round(ratio,2)
        gdp_ratios.append(ratio)
    return gdp_ratios
```

```python
#putting gdp ratios in new dataframe called gdp_ratios
d = {'Years': [2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010, 2009, 2008, 2007, 2006, 2005, 2004, 2003, 2002]}
gdp_ratios = pd.DataFrame(data=d)

gdp_ratios["White (lowest) gdp"]=get_gdp_ratio(all_income["White (lowest)"], all_income["GDP per capita"])
gdp_ratios["White (lowest) gdp"]=gdp_ratios["White (lowest) gdp"]

gdp_ratios["Black (lowest) gdp"]=get_gdp_ratio(all_income["Black (lowest)"], all_income["GDP per capita"])
gdp_ratios["Black (lowest) gdp"]=gdp_ratios["Black (lowest) gdp"]

gdp_ratios["Asian (lowest) gdp"]=get_gdp_ratio(all_income["Asian (lowest)"], all_income["GDP per capita"])
gdp_ratios["Asian (lowest) gdp"]=gdp_ratios["Asian (lowest) gdp"]

gdp_ratios["WANH (lowest) gdp"]=get_gdp_ratio(all_income["White Alone, Not Hispanic (lowest)"], all_income["GDP per capita"])
gdp_ratios["WANH (lowest) gdp"]=gdp_ratios["WANH (lowest) gdp"]

gdp_ratios["Hispanic (lowest) gdp"]=get_gdp_ratio(all_income["Hispanic (lowest)"], all_income["GDP per capita"])
gdp_ratios["Hispanic (lowest) gdp"]=gdp_ratios["Hispanic (lowest) gdp"]


gdp_ratios["White (highest) gdp"]=get_gdp_ratio(all_income["White (highest)"], all_income["GDP per capita"])
gdp_ratios["White (highest) gdp"]=gdp_ratios["White (highest) gdp"]

gdp_ratios["Black (highest) gdp"]=get_gdp_ratio(all_income["Black (highest)"], all_income["GDP per capita"])
gdp_ratios["Black (highest) gdp"]=gdp_ratios["Black (highest) gdp"]

gdp_ratios["Asian (highest) gdp"]=get_gdp_ratio(all_income["Asian (highest)"], all_income["GDP per capita"])
gdp_ratios["Asian (highest) gdp"]=gdp_ratios["Asian (highest) gdp"]

gdp_ratios["WANH (highest) gdp"]=get_gdp_ratio(all_income["White Alone, Not Hispanic (highest)"], all_income["GDP per capita"])
gdp_ratios["WANH (highest) gdp"]=gdp_ratios["WANH (highest) gdp"]

gdp_ratios["Hispanic (highest) gdp"]=get_gdp_ratio(all_income["Hispanic (highest)"], all_income["GDP per capita"])
gdp_ratios["Hispanic (highest) gdp"]=gdp_ratios["Hispanic (highest) gdp"]
```

Next, I want to see how each top 5% compared to their bottom 20% and if the ratio for that was proportional across racial groups. 

```python
#function to get ratios 
def get_ratios(column, highest):
    ratios = []
    for value in highest: 
        for i in column:
            ratio=int(value/i)
        ratios.append(ratio)
    return ratios
```

I computed the ratios over the years for each racial group and put them in a new dataframe called all_ratios.

```python
#creating and putting ratios for lowest and highest for each race in new dataframe all_ratios
d = {'Years': [2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010, 2009, 2008, 2007, 2006, 2005, 2004, 2003, 2002]}
all_ratios = pd.DataFrame(data=d)

# white
all_ratios["White ratio"]=get_ratios(all_income["White (lowest)"], all_income["White (highest)"])
all_ratios["White ratio"]=all_ratios["White ratio"]

# black
all_ratios["Black ratio"]=get_ratios(all_income["Black (lowest)"], all_income["Black (highest)"])
all_ratios["Black ratio"]=all_ratios["Black ratio"]


# asian
all_ratios["Asian ratio"]=get_ratios(all_income["Asian (lowest)"], all_income["Asian (highest)"])
all_ratios["Asian ratio"]=all_ratios["Asian ratio"]

# white alone not hispanic
all_ratios["WANH ratio"]=get_ratios(all_income["White Alone, Not Hispanic (lowest)"], all_income["White Alone, Not Hispanic (highest)"])
all_ratios["WANH ratio"]=all_ratios["WANH ratio"]


# hispanic
all_ratios["Hispanic ratio"]=get_ratios(all_income["Hispanic (lowest)"], all_income["Hispanic (highest)"])
all_ratios["Hispanic ratio"]=all_ratios["Hispanic ratio"]

all_ratios
```
